```python
import pickle
import os
import pandas as pd
```

```python
network = 'vgg16'
# network = 'res101'
output_path = './output/'+ network +'/ocid_test/faster_rcnn_10/'
_files = os.listdir(output_path)
files = [f for f in _files if os.path.isfile(os.path.join(output_path, f))]
assert len(files) == len(_files), "There is a directory in the output destination."
```

```python
report = []
for f in files:
    
    if f == 'detections.pkl':
        continue
        
    item = {}
    cls = f.replace('_pr.pkl', '')
    path = os.path.join(output_path, f)
    with open(path, 'rb') as pf:
        res = pickle.load(pf)
        pf.close()
    
    item['class'] = cls
    item['ap'] = res['ap']
    report.append(item)
```

```python
df_report = pd.DataFrame(report)
```

```python
df_report
```

```python
print ('map: %.4f' % (df_report.ap.mean()))
```

```python
df_report.to_csv(network + '_report.csv', index=False)
```

```python
with open('data/OCID-dataset/train.txt', 'r') as f:
    lines = f.readlines()
    f.close()
     
trainset = [line.strip() for line in lines]
```
